```python
# Load files.

datadir='/Users/AmandaE/Documents/MATLAB/projects/abide/abide60_pls/fdr_comparison/' # folder name
filename = 'plsconn_test_30.mat' # name of .mat file
matdata = scipy.io.loadmat(datadir+filename)
```

```python
fig=plt.figure(figsize=(9,8))

con_method = 'conn_test_xcorr_30' # name of adjacency matrix
title = 'Dynamic FC'
savename = datadir+con_method+'.jpg'

n_lines = size(find(matdata[con_method]!=0))/2

node_colors=['black']

plot_connectivity_circle(matdata[con_method],labels,n_lines=n_lines,
                         fig=fig,title=title,fontsize_title=16,
                         colormap='cool',facecolor='white',textcolor='black',
                         colorbar=0,
                         node_colors=node_colors,node_edgecolor=['white'])
    
fig.savefig(savename,dpi=300)
```
