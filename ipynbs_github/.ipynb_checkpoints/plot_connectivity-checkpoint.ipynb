```python
# Load files.

mypath='/Users/AmandaE/Documents/MATLAB/projects/abide/abide_pls/new_dosenbach/' # folder name
filename = 'test_pls_connmat.mat' # name of .mat file
matdata = scipy.io.loadmat(mypath+filename)
```

```python
fig=plt.figure(figsize=(9,8))

con_method = 'test_pls_connmat' # name of adjacency matrix
con = matdata[con_method]
title = 'Static FC'
savename = mypath+con_method+'.jpg'

con = con[0:90,0:90]
n_lines = size(find(con!=0))/2

node_colors=['black']

plot_connectivity_circle(con,labels,n_lines=90,
                         fig=fig,title=title,fontsize_title=16,
                         colormap='cool',facecolor='white',textcolor='black',
                         colorbar=0,node_colors=node_colors,node_edgecolor='white')
    
fig.savefig(savename,dpi=300)
```
